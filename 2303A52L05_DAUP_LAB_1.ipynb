b) Scatter plot with customizations

```python
import matplotlib.pyplot as plt

# Scatter plot between age and smoking
plt.scatter(data['age'], data['smoker'], marker='x')
plt.title('Scatter Plot between Age and Smoking')
plt.xlabel('Age')
plt.ylabel('Smoking (yes/no)')
plt.text(30, 'yes','.', fontsize=12, ha='right')
plt.show()
```

```python
x_column = data.columns[0]
y_column = data.columns[1]

plt.figure(figsize=(10, 6))
plt.scatter(data[x_column], data[y_column], c='green', marker='+', label='Data Points')
plt.title("Scatter Plot of {} vs {}".format(x_column, y_column))
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.legend()
```

c) Filter the data and update plot

```python
# Scatter plot between age (20 to 40) and smoking
filtered_df = data[(data['age'] >= 20) & (data['age'] <= 40)]
plt.scatter(filtered_df['age'], filtered_df['smoker'], marker='+')
plt.title('Scatter Plot between Age (20 to 40) and Smoking')
plt.xlabel('Age')
plt.ylabel('Smoking')
plt.show()
```

d) Expenses of smokers and non-smokers

```python
# Find expenses of smokers and non-smokers
smokers_expenses = data[data['smoker'] == 'yes']['expenses']
non_smokers_expenses = data[data['smoker'] == 'no']['expenses']

plt.boxplot([smokers_expenses, non_smokers_expenses], labels=['Smokers', 'Non-Smokers'])
plt.title('Expenses of Smokers vs Non-Smokers')
plt.ylabel('Expenses')
plt.show()
```

```python
x_column = data.columns[0]
y_column = data.columns[2]

plt.figure(figsize=(10, 6))
plt.scatter(data[x_column], data[y_column], c='red', marker='+', label='Data Points')
plt.title("Scatter Plot of {} vs {}".format(x_column, y_column))
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.legend()

for i in range(5):
    plt.text(data[x_column][i], data[y_column][i], str(i), fontsize=9)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Select numerical columns (excluding 'expenses')
numerical_cols = data.select_dtypes(include=['number']).columns.difference(['expenses'])

# Create scatter plots for each numerical column against 'expenses'
for col in numerical_cols:
    plt.figure(figsize=(8, 6))  # Adjust figure size as needed
    sns.scatterplot(x=col, y='expenses', data=data)
    plt.title(f'Scatter Plot: {col} vs. Expenses')
    plt.xlabel(col)
    plt.ylabel('Expenses')
    plt.show()
```

```python
# Select categorical columns
categorical_cols = data.select_dtypes(include=['object']).columns

# Create bar plots or box plots for each categorical column against 'expenses'
for col in categorical_cols:
    plt.figure(figsize=(8, 6))  # Adjust figure size as needed
    # Choose either bar plot or box plot based on your preference
    # sns.barplot(x=col, y='expenses', data=df)  # For bar plot
    sns.boxplot(x=col, y='expenses', data=data)  # For box plot
    plt.title(f'{col.capitalize()} vs. Expenses')
    plt.xlabel(col.capitalize())
    plt.ylabel('Expenses')
    plt.show()
```

```python
import pandas as pd

# Assuming your dataset is loaded into a pandas DataFrame called 'df'
# If not, replace 'df' with the actual name of your DataFrame
# df = pd.read_csv('your_dataset.csv')  # Replace 'your_dataset.csv' with the actual file path

# Group by 'sex' and calculate the sum of expenses for each group
gender_expenses = data.groupby('sex')['expenses'].sum()

# Print the results
print(gender_expenses)

# Determine which gender has more expenses
if gender_expenses['male'] > gender_expenses['female']:
    print("Males have more expenses.")
elif gender_expenses['female'] > gender_expenses['male']:
    print("Females have more expenses.")
else:
    print("Both genders have equal expenses.")
```

e) Number of smokers gender-wise

```python
# Number of smokers gender-wise
smoker_gender_count = data[data['smoker'] == 'yes'].groupby('sex').size()

smoker_gender_count.plot(kind='bar',color = 'brown')
plt.title('Number of Smokers by Gender')
plt.ylabel('Count')
plt.show()
```

NUMBER OF TEENAGE SMOKERS

```python
import pandas as pd

# Assuming your dataset is loaded into a pandas DataFrame called 'df'
# Replace 'your_dataset.csv' with the actual file path

# Filter for teenagers (age between 13 and 19) who are smokers
teenage_smokers = data[(data['age'] >= 13) & (data['age'] <= 19) & (data['smoker'] == 'yes')]

# Get the number of teenage smokers
num_teenage_smokers = len(teenage_smokers)

# Print the result
print("Number of teenage smokers:", num_teenage_smokers)
```
